 # TF-IDF Tutorial for Beginners

 In this notebook, we’ll go over what TF-IDF is and how to use it in Python to analyze text data. TF-IDF is a common technique in Natural Language Processing for transforming text into numerical form to analyze it or use it in machine learning.

 ### What is TF-IDF?

 **TF-IDF** stands for **Term Frequency-Inverse Document Frequency**. It’s a way to represent text data by assigning a numerical weight to each word in a document. Words that are more common in a document but rare in other documents get a higher weight, helping to highlight words that are important in one document but not in others.

 Here’s the formula for TF-IDF:

 **TF-IDF(term, document) = TF(term, document) * IDF(term)**

 Where:
 - **Term Frequency (TF)** measures how often a word appears in a document.
 - **Inverse Document Frequency (IDF)** reduces the importance of words that are common across many documents (like "the" or "is").

### Step 1: Importing Libraries

 We'll be using `TfidfVectorizer` from `sklearn`, a popular Python library for machine learning.




```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ## Step 2: Example Documents

# Let's create a few example documents to work with.

documents = [
    "This is a sample document about machine learning.",
    "Machine learning and data science are popular topics.",
    "Data science involves analyzing data and making predictions.",
    "This document is about analyzing data with machine learning.",
]
```

```python
# ## Step 3: Calculating TF-IDF

# Now, we'll use `TfidfVectorizer` to calculate TF-IDF for each word in each document.

# Create the TF-IDF vectorizer
vectorizer = TfidfVectorizer()
```

```python
# Fit and transform the documents
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
tfidf_matrix
```

```python
print(tfidf_matrix)
```

```python
# Convert the TF-IDF matrix into a pandas DataFrame for easy viewing
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
tfidf_df
```

```python
# ## Step 4: Interpreting TF-IDF Results

# In the output above, each row represents a document, and each column represents a word from our documents. The numbers in each cell represent the TF-IDF weight for that word in that document.

# - Higher values mean the word is more important in that document.
# - Words like "machine" and "learning" might have higher values because they're specific to the topic, whereas common words like "this" or "is" are downweighted.
```

```python
# Set up the figure size
plt.figure(figsize=(10, 6))

# Create the heatmap
sns.heatmap(tfidf_df, annot=True, cmap="YlGnBu", cbar_kws={'label': 'TF-IDF Score'}, fmt=".2f")

# Set labels for readability
plt.xlabel("Words")
plt.ylabel("Documents")
plt.title("TF-IDF Scores for Each Word in Each Document")
plt.show()
```

5. Many libraries such as sklearn have built-in toy datasets for practice. One famous one is the 20 Newsgroups dataset. Using the code below, import the newsgroups, and look at the data, to get a sense of what it is.

```python
# ## Expanding to a Larger Corpus: 20 Newsgroups Dataset

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Load the 20 Newsgroups dataset
newsgroups = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))

# Preview the dataset
print("Number of documents:", len(newsgroups.data))
print("Categories:", newsgroups.target_names)

# Display the first document as an example
print("\nSample Document:\n", newsgroups.data[0])
```

5 (cont) : Look at a few more entries, from other categories. what do you see?

```python
# Create the TF-IDF vectorizer with a limited vocabulary size for efficiency
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')

# Fit and transform the dataset
tfidf_matrix = vectorizer.fit_transform(newsgroups.data)

# Convert the TF-IDF matrix to a pandas DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Display the shape and a preview of the DataFrame
print("TF-IDF Matrix Shape:", tfidf_df.shape)
tfidf_df.head()
# to switch from data to code cell CTRL+MY
```

```python
# Add category labels to the TF-IDF DataFrame
tfidf_df['category'] = [newsgroups.target_names[i] for i in newsgroups.target]

# Group by category and calculate the mean TF-IDF score for each word
category_tfidf = tfidf_df.groupby('category').mean()

# Preview the average TF-IDF scores for each category
category_tfidf.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select top words by their average TF-IDF scores across categories
top_words = category_tfidf.mean().nlargest(20).index

# Filter the category_tfidf DataFrame to only include the top words
category_tfidf_top = category_tfidf[top_words]

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(category_tfidf_top, annot=True, fmt=".2f", cmap="YlGnBu", cbar_kws={'label': 'Average TF-IDF Score'})
plt.title("Top 20 Words by Average TF-IDF Score Across 20 Newsgroups Categories")
plt.xlabel("Top Words")
plt.ylabel("Categories")
plt.show()
```
